```python
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from kneed import KneeLocator
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

```python
df_1 = pd.read_csv('data_2d.csv')
df_2 = pd.read_csv('mnist.csv')

df_1
df_2
```

```python
from sklearn.cluster import KMeans
import numpy as np
from sklearn import decomposition


''' apply k-means algorithm for clustering for 'data_2d.csv' '''

X_1 = np.array(df_1)
kmeans = KMeans(n_clusters=2, random_state=0, n_init="auto").fit(X_1)
print(kmeans.labels_)
print(kmeans.cluster_centers_)

'''apply k-means algorithm for clustering for 'mnist.csv' '''

two_dt = np.array(df_2)
df = StandardScaler().fit_transform(two_dt)
X_2 = pd.DataFrame(df)

'''PCA algorithm to reduce data dimension to 2-dimensional version'''

pca = decomposition.PCA(n_components=2)
pca.fit(X_2)
kmeans = KMeans(n_clusters=2, random_state=0, n_init="auto").fit(X_2)
print(kmeans.labels_)
print(kmeans.cluster_centers_)
```

```python
'''scaled_features values scaled'''

scaler_1 = StandardScaler()
scaled_features_1 = scaler_1.fit_transform(X_1)
kmeans_1 = KMeans(
    init="random",
    n_clusters=3,
    n_init=10,
    max_iter=300,
    random_state=42
)
kmeans_1.fit(scaled_features_1)


scaler_2 = StandardScaler()
scaled_features_2 = scaler_2.fit_transform(X_2)
kmeans_2 = KMeans(
    init="random",
    n_clusters=3,
    n_init=10,
    max_iter=300,
    random_state=42
)
kmeans_2.fit(scaled_features_2)
```

```python
print(kmeans_1.inertia_)
print(kmeans_1.cluster_centers_)
print(kmeans_1.n_iter_)

print(kmeans_2.inertia_)
print(kmeans_2.cluster_centers_)
print(kmeans_2.n_iter_)
```

```python
'''To run the elbow method, run several k-means, and increment k with each iteration and write to SSE'''

kmeans_kwargs_1 = {
    "init": "random",
    "n_init": 10,
    "max_iter": 300,
    "random_state": 42,
}

sse_1 = []
for k in range(1, 11):
    '''A list holds the SSE values for each k'''
    kmeans_1 = KMeans(n_clusters=k, **kmeans_kwargs_1)
    kmeans_1.fit(scaled_features_1)
    sse_1.append(kmeans_1.inertia_)


kmeans_kwargs_2 = {
    "init": "random",
    "n_init": 10,
    "max_iter": 300,
    "random_state": 42,
}

sse_2 = []
for k in range(1, 11):
    '''A list holds the SSE values for each k'''
    kmeans_2 = KMeans(n_clusters=k, **kmeans_kwargs_2)
    kmeans_2.fit(scaled_features_2)
    sse_2.append(kmeans_2.inertia_)
```

```python
'''Visualization of clustering results'''

plt.style.use("fivethirtyeight")
plt.plot(range(1, 11), sse_1)
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()


plt.style.use("fivethirtyeight")
plt.plot(range(1, 11), sse_2)
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```

```python
'''Knee, for programmatic determination of the bending point'''

kl_1 = KneeLocator(range(1, 11), sse_1, curve="convex", direction="decreasing")
print(kl_1.elbow)
kl_2 = KneeLocator(range(1, 11), sse_2, curve="convex", direction="decreasing")
print(kl_2.elbow)
```

```python
'''silhouette coefficient'''

silhouette_coefficients_1 = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs_1)
    kmeans.fit(scaled_features_1)
    score = silhouette_score(scaled_features_1, kmeans.labels_)
    silhouette_coefficients_1.append(score)
    
silhouette_coefficients_2 = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs_2)
    kmeans.fit(scaled_features_2)
    score = silhouette_score(scaled_features_2, kmeans.labels_)
    silhouette_coefficients_2.append(score)
```

```python
'''Visualization of silhouette coefficient results'''

plt.style.use("fivethirtyeight")
plt.plot(range(2, 11), silhouette_coefficients_1)
plt.xticks(range(2, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Coefficient")
plt.show()

plt.style.use("fivethirtyeight")
plt.plot(range(2, 11), silhouette_coefficients_2)
plt.xticks(range(2, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Coefficient")
plt.show()
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(scaled_features_1)
kmeans_silhouette_1 = silhouette_score(scaled_features_1, kmeans.labels_).round(2)
print(kmeans_silhouette_1)

kmeans = KMeans(n_clusters=2)
kmeans.fit(scaled_features_2)
kmeans_silhouette_2 = silhouette_score(scaled_features_2, kmeans.labels_).round(2)
print(kmeans_silhouette_2)
```
